```python
import pandas as pd
```

## feestdagen dataset

```python
feestdagen_df = pd.read_csv("MLP_dataset.csv")
col = feestdagen_df.pop("matches")
feestdagen_df["matches"] = col # zodat dit de laatste kolom wordt voor iloc

# poi_id string maken -> hotonencoding
feestdagen_df["poi_id"] = feestdagen_df["poi_id"].astype(str)

feestdagen = ['2025-12-05', '2025-12-25', '2025-12-31', '2026-01-01']
feestdagen_df = feestdagen_df[feestdagen_df['date'].isin(feestdagen)]

feestdagen_df = feestdagen_df.drop(columns=["date"])

feestdagen_df.to_csv("feestdagen_dataset.csv", index=False)
```
